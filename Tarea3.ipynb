```python
#Importar librerias

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
#Cargar datos

np.random.seed(0)
xhoras_estudio = np.random.rand(100, 1) * 10
ycalificaciones = 2 * xhoras_estudio+ np.random.randn(100,1) * 2
```

```python
#Preprocesamiento de datos

# Creacion de regresion lineal
train_data, test_data, train_labels, test_labels = train_test_split(
    xhoras_estudio, ycalificaciones, test_size=0.33, random_state=50)

# Modelo de regresion lineal
model = LinearRegression()

# Entrenamiento del modelo
model.fit(xhoras_estudio, ycalificaciones)
```

```python
#Prueba del modelo

# Realizacion de predicciones
predictions = model.predict(test_data)
```

```python
#Calcula error cuadratico medio

mse = mean_squared_error(test_labels, predictions)
print(f"Error Cudratico Medio: {mse}")
```

```python
#Grafica del modelo de regresión lineal
#Colores y etiquetas de la grafica
plt.scatter(test_data, test_labels, color='black',
            label='Datos de Prueba')
plt.plot(test_data, predictions, color='red',
         label='Predicciones del Modelo')

# Area de etiquetación
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de Regresión Lineal')
plt.legend()
plt.show()
```
